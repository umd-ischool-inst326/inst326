```python
import pandas as pd
import re
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
def get_year(row):
    match = re.search(r".*\((\d+)\)", row["movie title"])
    return int(match[1])
```

```python
def get_title(row):
    match = re.search(r"^(.*\S)\s*\(\d{4}\)", row["movie title"])
    return match[1].strip()
```

```python
df["year"] = df.apply(get_year, axis=1)
df.head()
```

```python
df["title"] = df.apply(get_title, axis=1)
df.head()
```

```python
df2 = df[["movie id", "title", "year", "IMDb URL"]]
df2.head()
```

```python
dfr = pd.read_csv("ratings.csv")
```

```python
dfr.head()
```

```python
dfc = df2.merge(dfr, left_on="movie id", right_on="item id", how="inner").drop("item id", axis=1)
dfc.head()
```
